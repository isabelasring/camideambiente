# ðŸ“Š Ordenar posts por cantidad de comentarios

Organiza los posts de `postsCamilo.csv` del que tiene **mÃ¡s comentarios** al que tiene **menos**.

```python
import pandas as pd
import os

BASE_PATH = '/Users/geotrends/Desktop/scraping'
CSV_IN = os.path.join(BASE_PATH, 'postsCamilo.csv')
CSV_OUT = os.path.join(BASE_PATH, 'postsCamilo_ordenado.csv')  # Nuevo archivo ordenado

print(f'ðŸ“‚ Entrada: {CSV_IN}')
print(f'ðŸ“‚ Salida:  {CSV_OUT}')
```

```python
# Cargar CSV
df = pd.read_csv(CSV_IN, encoding='utf-8')

# Convertir commentsCount a numÃ©rico (valores invÃ¡lidos o -1 â†’ 0)
df['commentsCount'] = pd.to_numeric(df['commentsCount'], errors='coerce').fillna(0).astype(int)
df.loc[df['commentsCount'] < 0, 'commentsCount'] = 0

# Ordenar: mÃ¡s comentarios primero
df_ordenado = df.sort_values('commentsCount', ascending=False).reset_index(drop=True)

print(f'âœ… {len(df_ordenado):,} posts cargados')
print(f'\nðŸ“Š Top 5 con mÃ¡s comentarios:')
display(df_ordenado[['commentsCount', 'likesCount', 'timestamp', 'url']].head())
```

```python
# Guardar archivo ordenado
df_ordenado.to_csv(CSV_OUT, index=False, encoding='utf-8')
print(f'ðŸ’¾ Guardado: {CSV_OUT}')
print(f'   Del post con mÃ¡s comentarios ({df_ordenado["commentsCount"].iloc[0]}) al de menos ({df_ordenado["commentsCount"].iloc[-1]})')
```
