# ðŸ“„ Convertir perfilesSeguidores.json a CSV

Lee el archivo JSON con los perfiles completos (incluyendo los 16 faltantes) y lo exporta a formato CSV para mayor compatibilidad.

```python
import os
import json
import pandas as pd

# Rutas (ajustar en Colab si es necesario)
BASE_PATH = '/Users/geotrends/Desktop/scraping'
SCRAPERS_PATH = os.path.join(BASE_PATH, 'scrapers')

JSON_FILE = os.path.join(SCRAPERS_PATH, 'perfilesSeguidores.json')
CSV_FILE = os.path.join(SCRAPERS_PATH, 'perfilesSeguidores.csv')

print(f'ðŸ“‚ JSON: {JSON_FILE}')
print(f'ðŸ“‚ CSV destino: {CSV_FILE}')
```

```python
# Cargar JSON
with open(JSON_FILE, 'r', encoding='utf-8') as f:
    perfiles = json.load(f)

df = pd.DataFrame(perfiles)
print(f'âœ… Cargados {len(df):,} perfiles')
print(f'ðŸ“‹ Columnas: {list(df.columns)}')
df.head(3)
```

```python
# Exportar a CSV
df.to_csv(CSV_FILE, index=False, encoding='utf-8')
print(f'ðŸ’¾ Guardado: {CSV_FILE}')
print(f'   {len(df):,} filas, {len(df.columns)} columnas')
```
